# NorthStar Urban Mobility
## Step 1 — Data Exploration & Cleaning

```python
# pandas  → for reading and working with tables (like Excel)
# numpy   → for doing math on numbers
# matplotlib & seaborn → for making charts

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("All libraries loaded")
```

```python
import zipfile  # It is used to open and extract files from the ZIP folder.
import requests   # It is used to download the ZIP file directly from GitHub using the URL.
from io import BytesIO  # temporarily stores the downloaded ZIP file in memory so Python can read it without saving it manually.
```

```python
# Load all files directly from GitHub

zip_url = "https://raw.githubusercontent.com/1Shadha/NorthStar-Analytics/refs/heads/main/northstar_dataset.zip"

# Download the zip file
response = requests.get(zip_url)

# Open zip file in memory
zip_file = zipfile.ZipFile(BytesIO(response.content))

# Load CSV files from inside the zip
customers  = pd.read_csv(zip_file.open("northstar_dataset/customers.csv"))
orders     = pd.read_csv(zip_file.open("northstar_dataset/orders.csv"))
deliveries = pd.read_csv(zip_file.open("northstar_dataset/deliveries.csv"))
drivers    = pd.read_csv(zip_file.open("northstar_dataset/drivers.csv"))
vehicles   = pd.read_csv(zip_file.open("northstar_dataset/vehicles.csv"))
hubs       = pd.read_csv(zip_file.open("northstar_dataset/hubs.csv"))
incidents  = pd.read_csv(zip_file.open("northstar_dataset/incidents.csv"))
complaints = pd.read_csv(zip_file.open("northstar_dataset/complaints.csv"))
app_events = pd.read_csv(zip_file.open("northstar_dataset/app_events.csv"))

print("All files loaded from GitHub")
```

### 1. First Look at the Data
Looking at the first few rows of each table to understand the data structure.

```python
# .head() shows the first 5 rows of a table
# This helps us see what kind of data is inside

print("=== CUSTOMERS ===")
print(customers.head())

print("\n=== ORDERS ===")
print(orders.head())

print("\n=== DELIVERIES ===")
print(deliveries.head())

print("\n=== DRIVERS ===")
print(drivers.head())

print("\n=== VEHICLES ===")
print(vehicles.head())

print("\n=== HUBS ===")
print(hubs.head())

print("\n=== INCIDENTS ===")
print(incidents.head())

print("\n=== COMPLAINTS ===")
print(complaints.head())

print("\n=== APP EVENTS ===")
print(app_events.head())
```

```python
#To check the Shape of Each Table- which means how many rows and columns each table has.
datasets = {
    "customers": customers,
    "orders": orders,
    "deliveries": deliveries,
    "drivers": drivers,
    "vehicles": vehicles,
    "hubs": hubs,
    "incidents": incidents,
    "complaints": complaints,
    "app_events": app_events
}

print("Dataset shapes (rows, columns):\n")
for name, df in datasets.items():
    print(f"  {name:15} → {df.shape[0]} rows, {df.shape[1]} columns")
```

# CHECKING DATA QUALITY PROBLEMS

-Check for Missing Values

```python
# Check missing values one table at a time

print("Customers missing values:")
print(customers.isnull().sum())

print("\nOrders missing values:")
print(orders.isnull().sum())

print("\nDeliveries missing values:")
print(deliveries.isnull().sum())

print("\nDrivers missing values:")
print(drivers.isnull().sum())

print("\nVehicles missing values:")
print(vehicles.isnull().sum())

print("\nHubs missing values:")
print(hubs.isnull().sum())

print("\nIncidents missing values:")
print(incidents.isnull().sum())

print("\nComplaints missing values:")
print(complaints.isnull().sum())

print("\nApp Events missing values:")
print(app_events.isnull().sum())
```

-Check Inconsistent Spellings

```python
# Check zone columns for messy spellings

print("Customer home zones:")
print(customers['home_zone'].unique())

print("\nOrder pickup zones:")
print(orders['pickup_zone'].unique())

print("\nDriver base zones:")
print(drivers['base_zone'].unique())

print("\nDelivery status values:")
print(deliveries['delivery_status'].unique())
```

# FIXING DATA QUALITY ISSUES

-Fix Messy Spellings

```python
# Make all zone names consistent
# .str.title() makes first letter capital example: "NORTH" becomes "North"

customers['home_zone'] = customers['home_zone'].str.title()
orders['pickup_zone'] = orders['pickup_zone'].str.title()
orders['dropoff_zone'] = orders['dropoff_zone'].str.title()
drivers['base_zone'] = drivers['base_zone'].str.title()
vehicles['assigned_zone'] = vehicles['assigned_zone'].str.title()
app_events['zone_context'] = app_events['zone_context'].str.title()

print("Done! Spellings fixed.")
```

-Fix Missing Values

```python
# Fix ALL missing values

# --- CUSTOMERS ---
customers['preferred_channel'] = customers['preferred_channel'].fillna("Unknown")
customers['loyalty_score'] = customers['loyalty_score'].fillna(customers['loyalty_score'].median())

# --- ORDERS ---
orders['booking_channel'] = orders['booking_channel'].fillna("Unknown")

# --- DELIVERIES ---
deliveries['customer_rating_post_delivery'] = deliveries['customer_rating_post_delivery'].fillna(deliveries['customer_rating_post_delivery'].median())

# --- DRIVERS ---
drivers['training_score'] = drivers['training_score'].fillna(drivers['training_score'].median())

# --- VEHICLES ---
vehicles['battery_health_pct'] = vehicles['battery_health_pct'].fillna(vehicles['battery_health_pct'].median())

# --- INCIDENTS ---
incidents['resolved_hours'] = incidents['resolved_hours'].fillna(incidents['resolved_hours'].median())

# --- COMPLAINTS ---
complaints['compensation_amount'] = complaints['compensation_amount'].fillna(complaints['compensation_amount'].median())

print("All missing values fixed")
```

-Verify Cleaning

```python
# Check if missing values are gone now

print("Customers:")
print(customers.isnull().sum())

print("\nOrders:")
print(orders.isnull().sum())

print("\nDeliveries:")
print(deliveries.isnull().sum())

print("\nDrivers:")
print(drivers.isnull().sum())

print("\nVehicles:")
print(vehicles.isnull().sum())

print("\nIncidents:")
print(incidents.isnull().sum())

print("\nComplaints:")
print(complaints.isnull().sum())

print("\nApp Events:")
print(app_events.isnull().sum())
```

-Save cleaned data

```python
# Save all cleaned tables as new files

customers.to_csv("customers_clean.csv", index=False)
orders.to_csv("orders_clean.csv", index=False)
deliveries.to_csv("deliveries_clean.csv", index=False)
drivers.to_csv("drivers_clean.csv", index=False)
vehicles.to_csv("vehicles_clean.csv", index=False)
incidents.to_csv("incidents_clean.csv", index=False)
complaints.to_csv("complaints_clean.csv", index=False)
app_events.to_csv("app_events_clean.csv", index=False)

print("All clean files saved")
```
