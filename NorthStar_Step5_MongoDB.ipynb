-APP EVENTS — Nested Document Design

```python
app_events_collection = db["app_events"]
```

```python
# CREATE — Insert one nested app event
# Notice customer, device and location are NESTED inside!
# This is the NoSQL document design approach

nested_app_event = {
    "event_id": "AE00001",
    "event_type": "complaint_raised",
    "event_timestamp": "2024-08-09 03:25:00",
    "customer": {
        "customer_id": "C0488",
        "home_zone": "North",
        "customer_type": "Consumer"
    },
    "device": {
        "device_type": "Android",
        "session_id": "S19847",
        "api_latency_ms": 301
    },
    "location": {
        "zone_context": "North"
    },
    "order_id": "O00950",
    "success_flag": 1
}

app_events_collection.insert_one(nested_app_event)
```

```python
# READ — Show the nested document we just inserted

print("Nested document structure:")
print()
event = app_events_collection.find_one({"event_id": "AE00001"})
for key, value in event.items():
    print(f"  {key}: {value}")
```

```python
# CREATE — Insert all real app events from CSV

app_events_data = app_events_df.to_dict('records')
app_events_collection.insert_many(app_events_data)
```

```python
# READ — Show first 3 app events

print("Sample app events:")
for event in app_events_collection.find().limit(3):
    print(event)
```

```python
# UPDATE — Update nested device field
# This shows MongoDB can update fields INSIDE nested documents!

app_events_collection.update_one(
    {"event_id": "AE00001"},
    {"$set": {"device.api_latency_ms": 150}}
)

# Verify
verified = app_events_collection.find_one({"event_id": "AE00001"})
print(verified)
```

```python
# DELETE — Remove one app event

app_events_collection.delete_one({"event_id": "AE00001"})

# Verify
deleted = app_events_collection.find_one({"event_id": "AE00001"})
print(deleted)
```

# MongoDB Optimisation

Create Indexes

```python
# Create indexes to make searches faster

# Orders index
orders_collection.create_index("service_type")
print("Index created on orders!")

# Deliveries index
deliveries_collection.create_index("delivery_status")
print("Index created on deliveries!")

# Complaints index
complaints_collection.create_index("severity")
print("Index created on complaints!")
```

-Show indexes

```python
# Show indexes created

print("Orders indexes:")
for index in orders_collection.list_indexes():
    print(" ", index['name'])

print("\nDeliveries indexes:")
for index in deliveries_collection.list_indexes():
    print(" ", index['name'])

print("\nComplaints indexes:")
for index in complaints_collection.list_indexes():
    print(" ", index['name'])
```
